```python
from math import comb

def exact_p(k, n, t=7, deck=99):
    """
    Calculate the probability of having EXACTLY n target cards in a hand of t cards
    drawn from a deck of deck cards containing k target cards.
    
    Parameters:
    k (int): Number of target cards in the deck.
    n (int): Number of target cards in the hand.
    t (int): Number of cards drawn from the deck.
    
    Returns:
    float: Probability of having EXACTLY n target cards.
    """
    if n > k or t - n > deck - k or t > deck or n < 0:
        return 0  # Probability is zero if the combination is impossible
    
    numerator = comb(k, n) * comb(deck - k, t - n)
    denominator = comb(deck, t)
    return numerator / denominator

def p(k, n, t=7, deck=99):
    """
    Calculate the probability of having AT LEAST n target cards in a hand of t cards
    drawn from a deck of deck cards containing k target cards.
    
    Parameters:
    k (int): Number of target cards in the deck.
    n (int): Number of target cards in the hand.
    t (int): Number of cards drawn from the deck.
    
    Returns:
    float: Probability of having AT LEAST n target cards.
    """        
    return sum(exact_p(k, i, t, deck) for i in range(n, min(k, t) + 1))
```
